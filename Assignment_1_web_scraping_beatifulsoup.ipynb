Q.1) Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
def wikipedia(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    all_headers=[]
    heading = soup.find_all('span',class_ = 'mw-headline')
    for i in heading:
        all_headers.append(i.text)
        
        
    import pandas as pd
    wiki=pd.DataFrame()
    wiki['Heading']=all_headers
    return(wiki)
```

```python
lists = wikipedia("https://en.wikipedia.org/wiki/Main_Page")
lists
```

Q.2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
def Top_100_Movies(url):
    MovieName=[]
    MovieRating=[]
    YearOfRelease=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    mname=soup.find_all('td',class_="titleColumn")
    for i in mname:
        MovieName.append(i.find('a').get_text())
    mrating=soup.find_all('td',class_="ratingColumn imdbRating")
    for i in mrating:
        MovieRating.append(i.text.replace('\n',''))
    myor=soup.find_all('span',class_="secondaryInfo")
    for i in myor:
        YearOfRelease.append(i.text.replace('(','').replace(')',''))
    import pandas as pd
    imdb=pd.DataFrame({})
    imdb['MovieName']=MovieName
    imdb['MovieRating']=MovieRating
    imdb['YearOfRelease']=YearOfRelease
    imdb=imdb[0:100]
    print(imdb)
    imdb.to_csv('Top_100_movies.csv')
```

```python
lists = Top_100_Movies('https://www.imdb.com/chart/top')
lists
```

Q.4) Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
def BookPage(url):
    BookName=[]
    AuthorName=[]
    Genre=[]
    Review=[]
    BookPage=requests.get(url)
    soup=BeautifulSoup(BookPage.content,'html.parser')
    bnames=soup.find_all('h4',class_='italic')
    for i in bnames:
        BookName.append(i.find('a').text)
    bauthors=soup.find_all('p',class_='sans bold')
    for i in bauthors:
        AuthorName.append(i.text.replace('\n',''))
    bgenres=soup.find_all('p',class_='genre-links hidden-phone')
    for i in bgenres:
        Genre.append(i.find('a').text)
    breviews=soup.find_all('p',class_='excerpt')
    for i in breviews:
        Review.append(i.text.replace('\n',''))
    import pandas as pd
    books=pd.DataFrame()
    books['Name of the Book']=BookName
    books['Name of Author']=AuthorName
    books['Genres']=Genre
    books['Reviews']=Review
    return(books)
```

```python
lists = BookPage("https://bookpage.com/")
lists[0:5]
```

Q.5) Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
def OneDayTeams(url):
    Team=[]
    MatchPlayed=[]
    PointsEarned=[]
    Rating=[]
    
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
      
    TeamsName=soup.find_all('span',class_='u-hide-phablet')
    for i in TeamsName:
        Team.append(i.text)
    Team=Team[:10]
      
    first_pos=soup.find_all('td',class_='rankings-block__banner--matches')
    
    #For points also first and other positions have differnt format
    #for first position
    for i in first_pos:
        MatchPlayed.append(i.text)

               
    # For other positions
    other_pos=soup.find_all('td',class_='table-body__cell u-center-text')
    for i in range(0,len(other_pos),2):
        MatchPlayed.append(other_pos[i].text)
    MatchPlayed=MatchPlayed[:10]


    
    # For points also first and other positions have differnt format
    point_first=soup.find_all('td',class_='rankings-block__banner--points')
    
    for i in point_first:
        PointsEarned.append(i.text)
        
    
    point_others=soup.find_all('td',class_='table-body__cell u-center-text')
    for i in range(1,len(point_others),2):
        PointsEarned.append(point_others[i].text)
    PointsEarned=PointsEarned[:10]
    
    # For ratings also first and other positions have differnt format
    rating_first=soup.find_all('td',class_='rankings-block__banner--rating u-text-right')
    for i in rating_first:
        Rating.append(i.text.replace('\n','').replace(' ',''))
    
    rating_others=soup.find_all('td',class_='table-body__cell u-text-right rating')
    for i in rating_others:
        Rating.append(i.text.replace('\n','').replace(' ',''))
    Rating=Rating[:10]
    
    
    import pandas as pd
    one_day=pd.DataFrame()
    one_day['teams']=Team
    one_day['matches']=MatchPlayed
    one_day['points']=PointsEarned
    one_day['ratings']=Rating
    return(one_day)
```

```python
lists =OneDayTeams('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
lists
```

ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
def top_10_batsman(url):
    Batsman=[]
    Team=[]
    Rating=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    #extracting the top 10 batsman name
    # First Row is in different format
    first_batsman=soup.find_all('div',class_='rankings-block__banner--name-large')
    for i in first_batsman:
        Batsman.append(i.text)
    
    #other rows are in different format 
    other_batsman=soup.find_all('td',class_='table-body__cell rankings-table__name name')
    for i in other_batsman:
        Batsman.append(i.find('a').text)
    Batsman=Batsman[:10]
    
    #extracting the team name
    # First Row is in differeent format
    first_team=soup.find_all('div',class_='rankings-block__banner--nationality')
    for i in first_team:
        Team.append(i.text.replace('\n',''))
    
    # other rows are in different format
    other_team=soup.find_all('span',class_='table-body__logo-text')
    for i in other_team:
        Team.append(i.text.replace('\n',''))
    Team=Team[:10]
    
    #extracting the rating of players
    # First Row is in different format
    first_rating=soup.find_all('div',class_='rankings-block__banner--rating')
    for i in first_rating:
        Rating.append(i.text)
    # Next 9 rows are in different format
    other_rating=soup.find_all('td',class_='table-body__cell rating')
    for i in other_rating:
        Rating.append(i.text)
    Rating=Rating[:10]
    
    
    import pandas as pd
    batsmen=pd.DataFrame()
    batsmen['players']=Batsman
    batsmen['teams']=Team
    batsmen['ratings']=Rating
    return(batsmen)
```

```python
lists =top_10_batsman("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
lists
```

Q.7) Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def amazon(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    Product=[] 
    Price=[]
    Rating=[]
    Image=[]

    item= soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')
    for i in item:
        Product.append(i.text)
    Product=Product[0:5]
 
    
    price= soup.find_all('span',class_='a-price-whole')
    for i in price:
        Price.append(i.text)
    Price=Price[0:5]
     
    rating= soup.find_all('span',class_='a-icon-alt')
    for i in rating:
        Rating.append(i.text)
    Rating=Rating[0:5]
 
    image= soup.find_all('div', class_='a-section aok-relative s-image-fixed-height')
    for i in image:
        Image.append(i.find('img'))
    image=image[0:5] 
        
        
    import pandas as pd
    Mobiles=pd.DataFrame()
    Mobiles['Product']=Product
    Mobiles['Price']=Price
    Mobiles['Image']=Image[0:5]
    Mobiles['Rating']=Rating
    return(Mobiles)
```

```python
lists = amazon('https://www.amazon.in/Mobile-Phone-Under-20000-Rupees/s?k=Mobile+Phone+Under+20000+Rupees')
lists
```

Q.8)Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
def weather(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    periods=[]
    p=soup.find_all('p',class_='period-name')
    for i in p:
        periods.append(i.text)
    
    short_descriptions=[]
    s_d=soup.find_all('p',class_='short-desc')
    for i in s_d:
        short_descriptions.append(i.text)
        
    Temp=[]
    t=soup.find_all('p',class_='temp')
    for i in t:
        Temp.append(i.text)
        
    descriptions=[]
    des=soup.find_all('div',class_='col-sm-10 forecast-text')
    for i in des:
        descriptions.append(i.text)
    descriptions=descriptions[:9]
    
    import pandas as pd
    climate=pd.DataFrame()
    climate['Periods']=periods
    climate['Short_Descriptions']=short_descriptions
    climate['Temperature']=Temp
    climate['Descriptions']=descriptions
    return(climate)
```

```python
lists =weather('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.X80BdlUzbIU')
lists 
```
